```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/gabriel.pereira/Documents/Gabriel/SANTANDER CODERS/Introdução a Data Science/Criando graficos/TITLE,RELEASE_YEAR,SCORE,NUMBER_OF_.CSV")
```

```python
df.head()
```

```python
eixo_x = df['MAIN_GENRE'].value_counts().keys()
eixo_y = df['MAIN_GENRE'].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)

## adiciona titulo
plt.title('Best Movies for Genre')

## rotaciona a legenenda do eixo x
plt.xticks(rotation=45)

##adiciona label eixo x
plt.xlabel('Movies Genre')
plt.ylabel('Movies Quantity')


plt.show()
```

```python
eixo_x = df['RELEASE_YEAR'].value_counts().sort_index().keys()
eixo_y = df['RELEASE_YEAR'].value_counts().sort_index().values
```

```python
plt.plot(eixo_x, eixo_y)

## adiciona titulo
plt.title('Movies for Release Year')


##adiciona label eixo x
plt.xlabel('Release Year')
plt.ylabel('Movies Quantity')
```
